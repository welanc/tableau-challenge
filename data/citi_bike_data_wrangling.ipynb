```python
def week_data(dataframe):
    # drop na values from the DataFrame
    df_clean = dataframe.dropna()
    
    # extract the year and month from the starttime
    first_row = pd.Series(df_clean["starttime"][0])
    year_month = first_row.str.split("-",n=2).str[0] + "-" + first_row.str.split("-",n=2).str[1]
    
    # change starttime and stoptime series to date-time format
    df_clean["starttime"] = df_clean["starttime"].astype('datetime64[ns]')
    df_clean["stoptime"] = df_clean["stoptime"].astype('datetime64[ns]')
    
    # set "start date" as 7th day of the month
    start_date = dt.datetime.strptime(f"{year_month[0]}-07", "%Y-%m-%d")

    # set "end date" as 14th day of the month
    end_date = dt.datetime.strptime(f"{year_month[0]}-14", "%Y-%m-%d")
    
    # filter by date range
    after_start_date = df_clean["starttime"] >= start_date
    before_end_date = df_clean["starttime"] <= end_date
    between_two_dates = after_start_date & before_end_date
    
    # loc method to return a filtered DataFrame
    filtered_dates = df_clean.loc[between_two_dates]    
    filtered_dates["gender"] = filtered_dates["gender"].replace([0,1,2],["Unknown","Male","Female"])
    
    # return filtered DataFrame
    return filtered_dates
```

```python
feb_14_week = week_data(feb_14_df)
feb_14_week
```

```python
aug_14_week = week_data(aug_14_df)
aug_14_week
```

```python
feb_19_week = week_data(feb_19_df)
feb_19_week
```

```python
aug_19_week = week_data(aug_19_df)
aug_19_week
```

```python
weeks_df = [feb_14_week, aug_14_week, feb_19_week, aug_19_week]
```

```python
# Combine all four DataFrames into one using .concat() method
combined_df = pd.concat(weeks_df)
combined_df
```

```python
# reset the index
combined_df.reset_index(drop=True, inplace=True)
```

```python
# duplicate each row
dup_df = combined_df.reindex(combined_df.index.repeat(2)).reset_index(drop=True)
dup_df
```

```python
# get duplicated rows for loc searching
dups = dup_df.duplicated()
dups
```

```python
# fill all latitude and longitude rows with start lat,lons
dup_df["latitude"] = dup_df["start station latitude"]
dup_df["longitude"] = dup_df["start station longitude"]
```

```python
# fill all duplicated rows with end lat,lons
dup_df.loc[dups,"latitude"] = dup_df["end station latitude"]
dup_df.loc[dups,"longitude"] = dup_df["end station longitude"]
```

```python
# add new series called "location_name" and 
# fill with start station and end station, 
# to be used as an "ID" column to link 
# pairs of rows and their start and end latitude/longitude
dup_df["location_name"] = dup_df["start station name"] + "_" + dup_df["end station name"]
dup_df
```

```python
# series/columns where duplicated data is unnecessary
wipe_data = ['tripduration', 'starttime', 'stoptime', 'start station id',
       'start station name', 'start station latitude',
       'start station longitude', 'end station id', 'end station name',
       'end station latitude', 'end station longitude', 'usertype',
       'birth year', 'gender']
```

```python
# remove duplicated data from duplicated rows to avoid double counting
dup_df.loc[dups, wipe_data] = None
```
